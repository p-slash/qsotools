```python
import numpy as np
import matplotlib.pyplot as plt
import qsotools.specops as so
```

```python
# Plot for two different resolving powers
R1 = 35000
R2 = 70000
dv = 2.6
k  = np.logspace(-4, 0, 100)

plt.semilogx(k, so.getSpectResWindow(k, R1, dv), label="R=35k")
plt.semilogx(k, so.getSpectResWindow(k, R2, dv), label="R=70k")
plt.legend()
plt.xlabel(r"$k$ [s km$^{-1}$]")
plt.ylabel(r"$W(k; R, dv)$")
plt.grid(True, "major")
plt.ylim(0.8, 1.01)
```
