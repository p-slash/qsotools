```python
import numpy as np
import matplotlib.pyplot as plt
import qsotools.specops as so
```

```python
# Plot for two different resolving powers
R1 = 35000
R2 = 70000
dv = 2.6
k  = np.logspace(-4, 0, 100)

plt.semilogx(k, so.getSpectResWindow(k, R1, dv)**2, label="R=35k")
plt.semilogx(k, so.getSpectResWindow(k, R2, dv)**2, label="R=70k")
plt.legend()
plt.xlabel(r"$k$ [s km$^{-1}$]")
plt.ylabel(r"$W(k; R, dv)$")
plt.grid(True, "major")
plt.ylim(0.90, 1.01)
```

```python
# Plot for two different resolving powers
R1 = 70000
R2 = 70000
dv1 = 2.6
dv2 = 1.3
k  = np.logspace(-4, 0, 100)

plt.semilogx(k, so.getSpectResWindow(k, R1, dv1)**2/so.getSpectResWindow(k, R1, 2.0)**2)
plt.semilogx(k, so.getSpectResWindow(k, R2, dv2)**2/so.getSpectResWindow(k, R2, 2.0)**2)
plt.xlabel(r"$k$ [s km$^{-1}$]")
plt.grid(True, "major")
plt.axvline(0.3, c='k')
plt.ylim(0.94, 1.04)
```
